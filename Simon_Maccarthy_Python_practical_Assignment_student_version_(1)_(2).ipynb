> 💡**HINT**💡

**Function specifications**

*Arguments:*
- **matrix** `(numpy.ndarray)` $\rightarrow$ input matrix to be used for computing the sum.

*Return:*
- **sum** `(int)` $\rightarrow$  return the sum of the two diagonals.


````python
### START FUNCTION
def sum_of_diagonals(matrix):
    n = matrix.shape[0]  # Get the dimension of the square matrix
    diagonal_sum = 0

    # Sum main diagonal (top-left to bottom-right)
    for i in range(n):
        diagonal_sum += matrix[i, i]

    # Sum anti-diagonal (top-right to bottom-left)
    for i in range(n):
        diagonal_sum += matrix[i, n - 1 - i]

    # If n is odd, the center element was added twice, so subtract it once
    if n % 2 == 1:
        center_index = n // 2
        diagonal_sum -= matrix[center_index, center_index]

    return diagonal_sum
### END FUNCTION
````

Write a function that computes the ratio of vowels vs. consonants ($\frac{vowels}{consonants}$) in a given sentence. The ratio should be given as a floating-point number rounded off to two decimal places. When writing your function, be sure to appropriately cater for the following punctuation marks:
- apostrophes (')
- quotations (")
- full stops (.)
- commas (,)
- exclamations (!)  
- question marks (?)
- colons (:)
- semicolons (;)

*Arguments:*
- **sentence** `(string)` $\rightarrow$ sentence required to compute the ratio.

*Return:*
- **ratio** `(float)` $\rightarrow$ a ratio that describes the number of vowels to consonants: vowels/consonants.

> 💡**HINT**💡
>
>Remember to cater for the white spaces that appear in your sentence.

````python
def vowel_consonant_ratio(sentence):
    vowels = set("aeiouAEIOU")
    vowel_count = 0
    consonant_count = 0

    for ch in sentence:
        if ch.isalpha():  # only consider letters
            if ch in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    if consonant_count == 0:  # avoid division by zero
        return 0.0 # Return as float to match expected type of ratio

    ratio = vowel_count / consonant_count
    return round(ratio, 2)
````
