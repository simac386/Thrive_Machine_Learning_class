 What is Supervised Learning?

Supervised Learning is a type of Machine Learning where:

We have input data (features)

We also have the correct output (label/target)

The model learns the relationship between inputs and outputs

👉 In this dataset:

Inputs → Study_Hours, Attendance_Percentage, Previous_Score

Output → Exam_Marks

Since we already know the exam marks, the model can learn from them.


What is Regression?
Ans: Regression is a type of supervised learning that predicts a
continuous target variable based on input features.

 Key Idea: Find a relationship between the input variables (features)
and the target variable.

 Examples:
• Predicting house prices based on size, location, and number of bedrooms.
• Predicting the weight of a person based on their height and age.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns
```

Now we will load some dataset

```python
df = pd.read_csv('/content/Simon Maccarthy - supervised_linear_regression_dataset.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=["Attendance_Percentage", "Previous_Score"])
```

```python
df.head(10)
```

Define X (feature) and y (target)

```python
X = df[["Study_Hours"]]
y = df["Exam_Marks"]
```

# Create model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```


# Train model

```python
model.fit(X, y)
```

# Predictions

```python
predictions = model.predict(X)
```

# Print equation

```python
print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```

# Visualization

```python
plt.scatter(X, y, color='blue', label="Actual Marks")
plt.plot(X, predictions, color='red', label="Regression Line")
plt.xlabel("Study Hours")
plt.ylabel("Exam Marks")
plt.title("Study Hours vs Exam Marks")
plt.legend()
plt.show()
```

Model Training Explanation
When we run:

```python
model.fit(X, y)
```

The model learns the best-fit line:

Exam_Marks=m×Study_Hours+b

Where:
m=slope
b=intercept

From this dataset, the relationship is almost perfectly linear:
As Study Hours increase, Exams Marks increase

Visualization
(Study Hours vs Exam Marks)
                           
   🔵 Blue dots: Actual data

🔴 Red line: Predicted regression line.
The line goes upward, showing a strong positive relationship.

This means more study hours Higher exams

**Conclusion**:

From the analysis:
